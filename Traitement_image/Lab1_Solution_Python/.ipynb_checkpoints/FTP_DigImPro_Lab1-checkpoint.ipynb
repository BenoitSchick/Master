2. Show image information

```python
def show_image_info_numpy(prompt, image):
	try:
		#Color image
		h, w, c = image.shape
	except:
		#Gray scale image
		h, w = image.shape
		c=1

	datatype = image.dtype

	# getsizeof donne un nombre de bytes par pixel gigantesque (96) pour uint8
	#size = h * w * c * sys.getsizeof(datatype)
	size = h * w * c   #1 byte par pixel généralement
	print("%s: " %prompt)
	print("DEFINITION :  width: %d, height: %d, channels: %d" % (w, h, c))
	print("SIZE       :  datatype: %s, size: %d[bytes] %d[KB]" % (datatype, size, size/1024 ))

show_image_info_numpy("house.pgm: ", img_house)
show_image_info_numpy("lena10.pgm: ", img_lena)
show_image_info_numpy("Maison.pgm: ", img_maison)
show_image_info_numpy("Mandrill.bmp: ", img_mandrill)
```

## Arithmetic operations on images

3a. Addition of two images: `house + lena10`

```python
# 1) with +  
sum_with_plus = img_house + img_lena		# overflow --> modulo 256
plt.subplot(1,2,1)
plt.title("img_house + img_lena")
plt.imshow(sum_with_plus)
plt.axis("off")

# 2) with cv2.add
sum_with_cv2_add= cv2.add(img_house,img_lena)# overflow --> saturation (>255:white)
plt.subplot(1,2,2)
plt.title("cv2.add(img_house,img_lena)")
plt.imshow(sum_with_cv2_add)
result = plt.axis("off")
```

3b. Subtraction of two images: `house - lena10`

```python

# 1) with -  
diff_with_minus = img_house + img_lena		# underflow --> modulo 256
plt.subplot(1,2,1)
plt.title("img_house - img_lena")
plt.imshow(diff_with_minus)
plt.axis("off")

# 2) with cv2.subtract
diff_with_cv2_subtract= cv2.subtract(img_house,img_lena)# overflow --> saturation (<0: black)
plt.subplot(1,2,2)
plt.title("cv2.subtract(img_house,img_lena)")
plt.imshow(diff_with_cv2_subtract)
success=plt.axis("off")
```

4. Compensation of inhomogeneous illumination

thanks to the background image with uneven illumination of a uniform surface, it is possible to compensate and improve the text image quality

```python
img_text_shadow = cv2.imread("shadowy_text.png")
img_background = cv2.imread("inhomogeneous_background.png")

plt.rcParams["figure.figsize"] = (16,8)
plt.subplot(1,2,1)
x=plt.imshow(img_background)
plt.axis("off")
plt.subplot(1,2,2)
x=plt.imshow(img_text_shadow)
plt.axis("off")

# with subtraction
plt.figure()
plt.subplot(1,2,1)
plt.title("background subtraction")
img_compensated = cv2.subtract(img_background, img_text_shadow)
x=plt.imshow(2*img_compensated)
plt.axis("off")

# with division
plt.subplot(1,2,2)
plt.title("background division")
img_compensated = cv2.divide(img_background, img_text_shadow)
scale=int(255/ np.max(img_compensated))
x=plt.imshow(img_compensated * scale)
plt.axis("off")
```

## Image Regions

5. Extract the boundaries of white regions from a binary image

```python
img_binary = cv2.imread('test5.png',0)
print(img_binary.shape)
cv2.imshow('test5.png',img_binary)
cv2.waitKey(2500)
cv2.destroyAllWindows()

def extract_boundary(image_source):
	image_border = np.zeros(image_source.shape, np.uint8)
	count = 0
	for i in range(1,image_source.shape[0]-1):
		for j in range(1,image_source.shape[1]-1):
			if image_source[i,j] == 255:
				if image_source[i-1,j] == 0 or image_source[i+1,j] == 0 or image_source[i,j-1] == 0 or image_source[i,j+1] == 0:
					image_border[i,j] = 255
					count += 1
	print("Number of pixels in the border: ", count)
	return image_border

img_boundary = extract_boundary(img_binary)
plt.rcParams["figure.figsize"] = (20,10)
plt.subplot(1,3,1)
plt.title("test5.png (binary)")
plt.imshow(img_binary,cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("test5.png, V4-boundary")
plt.imshow(img_boundary,cmap='gray')
plt.axis("off")

img_inner_region = cv2.bitwise_xor(img_binary, img_boundary)	

plt.subplot(1,3,3)
plt.title("test5.png, inner region")
plt.imshow(img_inner_region,cmap='gray')
plt.axis("off")
```
