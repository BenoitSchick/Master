# Exercice 4 Data visualization
To train a bit yourself, we propose you to do a visualisation workout with the Iris dataset
https://en.wikipedia.org/wiki/Iris_flower_data_set. Download the file iris.txt from
moodle or from http://www.statlab.uni-heidelberg.de/data/iris/. Put it in a Python
data structure and attempt to reproduce a plot close to the one in Figure 4. Advice : start
by working with individual plots and then move to a grid of plots with subplot or take the
shortcut of using the library pandas.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

iris = pd.read_csv("iris.txt", sep="\t")
#print(iris)
#print(iris['Sepal.Length'])

sepal_length_setosa = []
sepal_width_setosa = []
petal_width_setosa = []
petal_length_setosa = []

sepal_length_versicolor = []
sepal_width_versicolor = []
petal_width_versicolor = []
petal_length_versicolor = []

sepal_length_virginica = []
sepal_width_virginica = []
petal_width_virginica = []
petal_length_virginica = []


for sp, sepal_length, sepal_width, petal_length, petal_width in zip(iris['Species'], iris['Sepal.Length'], iris['Sepal.Width'], iris['Petal.Length'], iris['Petal.Width']):
    if sp == 'setosa':
        sepal_length_setosa.append(sepal_length)
        sepal_width_setosa.append(sepal_width)
        petal_length_setosa.append(petal_length)
        petal_width_setosa.append(petal_width)

    if sp == 'versicolor':
        sepal_length_versicolor.append(sepal_length)
        sepal_width_versicolor.append(sepal_width)
        petal_width_versicolor.append(petal_width)
        petal_length_versicolor.append(petal_length)

    if sp == 'virginica':
        sepal_length_virginica.append(sepal_length)
        sepal_width_virginica.append(sepal_width)
        petal_width_virginica.append(petal_width)
        petal_length_virginica.append(petal_length)

fig, axes = plt.subplots(4, 4, figsize=(12, 10))
axes = axes.flatten()


# axes[0] : histogramme Sepal.Length
axes[0].hist(iris['Sepal.Length'], bins=10)

# axes[1] : scatter Sepal.Width vs Sepal.Length
axes[1].scatter(sepal_width_setosa, sepal_length_setosa, s=20, c='red', label='setosa')
axes[1].scatter(sepal_width_versicolor, sepal_length_versicolor, s=20, c='green', label='versicolor')
axes[1].scatter(sepal_width_virginica, sepal_length_virginica, s=20, c='blue', label='virginica')

# axes[2] : scatter Petal.Length vs Sepal.Length
axes[2].scatter(petal_length_setosa, sepal_length_setosa, s=20, c='red')
axes[2].scatter(petal_length_versicolor, sepal_length_versicolor, s=20, c='green')
axes[2].scatter(petal_length_virginica, sepal_length_virginica, s=20, c='blue')


# axes[3] : scatter Petal.Width vs Sepal.Length
axes[3].scatter(petal_width_setosa, sepal_length_setosa, s=20, c='red')
axes[3].scatter(petal_width_versicolor, sepal_length_versicolor, s=20, c='green')
axes[3].scatter(petal_width_virginica, sepal_length_virginica, s=20, c='blue')


# axes[4] : scatter Sepal.Length vs Sepal.Width
axes[4].scatter(sepal_length_setosa, sepal_width_setosa, s=20, c='red')
axes[4].scatter(sepal_length_versicolor, sepal_width_versicolor, s=20, c='green')
axes[4].scatter(sepal_length_virginica, sepal_width_virginica, s=20, c='blue')


# axes[5] : histogramme Sepal.Width
axes[5].hist(iris['Sepal.Width'], bins=10)


# axes[6] : scatter Petal.Length vs Sepal.Width
axes[6].scatter(petal_length_setosa, sepal_width_setosa, s=20, c='red')
axes[6].scatter(petal_length_versicolor, sepal_width_versicolor, s=20, c='green')
axes[6].scatter(petal_length_virginica, sepal_width_virginica, s=20, c='blue')


# axes[7] : scatter Petal.Width vs Sepal.Width
axes[7].scatter(petal_width_setosa, sepal_width_setosa, s=20, c='red')
axes[7].scatter(petal_width_versicolor, sepal_width_versicolor, s=20, c='green')
axes[7].scatter(petal_width_virginica, sepal_width_virginica, s=20, c='blue')


# axes[8] : scatter Sepal.Length vs Petal.Length
axes[8].scatter(sepal_length_setosa, petal_length_setosa, s=20, c='red')
axes[8].scatter(sepal_length_versicolor, petal_length_versicolor, s=20, c='green')
axes[8].scatter(sepal_length_virginica, petal_length_virginica, s=20, c='blue')


# axes[9] : scatter Sepal.Width vs Petal.Length
axes[9].scatter(sepal_width_setosa, petal_length_setosa, s=20, c='red')
axes[9].scatter(sepal_width_versicolor, petal_length_versicolor, s=20, c='green')
axes[9].scatter(sepal_width_virginica, petal_length_virginica, s=20, c='blue')


# axes[10] : histogramme Petal.Length
axes[10].hist(iris['Petal.Length'], bins=10)


# axes[11] : scatter Petal.Width vs Petal.Length
axes[11].scatter(petal_width_setosa, petal_length_setosa, s=20, c='red')
axes[11].scatter(petal_width_versicolor, petal_length_versicolor, s=20, c='green')
axes[11].scatter(petal_width_virginica, petal_length_virginica, s=20, c='blue')


# axes[12] : scatter Sepal.Length vs Petal.Width
axes[12].scatter(sepal_length_setosa, petal_width_setosa, s=20, c='red')
axes[12].scatter(sepal_length_versicolor, petal_width_versicolor, s=20, c='green')
axes[12].scatter(sepal_length_virginica, petal_width_virginica, s=20, c='blue')


# axes[13] : scatter Sepal.Width vs Petal.Width
axes[13].scatter(sepal_width_setosa, petal_width_setosa, s=20, c='red')
axes[13].scatter(sepal_width_versicolor, petal_width_versicolor, s=20, c='green')
axes[13].scatter(sepal_width_virginica, petal_width_virginica, s=20, c='blue')


# axes[14] : scatter Petal.Length vs Petal.Width
axes[14].scatter(petal_length_setosa, petal_width_setosa, s=20, c='red')
axes[14].scatter(petal_length_versicolor, petal_width_versicolor, s=20, c='green')
axes[14].scatter(petal_length_virginica, petal_width_virginica, s=20, c='blue')


# axes[15] : histogramme Petal.Width
axes[15].hist(iris['Petal.Width'], bins=10)


labels = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]

for i in range(0,16):
    if i % 4 == 0 or i == 0:
        axes[i].set_ylabel(labels[i//4])
        if i == 12:
            axes[i].set_xlabel(labels[i-12])
        else:
            axes[i].set_xticks([])
            
    elif i==15 or i==14 or i==13:
        axes[i].set_xlabel(labels[i-12])
        axes[i].set_yticks([])
    else :    
        axes[i].set_xticks([])
        axes[i].set_yticks([])


#plt.xlabel('')
#plt.ylabel('')
plt.show()
```

    - **Unsupervised** : Pas de label qui classe à quel segment appartient chaque client. On doit donc trouver/former des groupes de clients similaires avec des algo de clustering.
- d) Given a dataset of patients diagnosed as either having glaucoma or not, learn to classify new patients as having glaucoma or not.
    - **Supervised** : Les données possedent des labels et il faut entrainer un algo capable de prédire le diagnostic pour un nouveau patient --> probleme de classification.

## 2) Classification vs. regression systems 

Can we transform a regression problem into a classification problem ? What would be the
benefits of doing so ?
- Oui... un regression problem prédit une valeur continue. On peut le transformer en un probleme de classification en discrétisant la sortie continue en classes (ex: regression --> prédit le prix exact d'une maison, classification --> prédit un interval de prix). Modèle plus simple car il a un nombre fini de classes (ce qui est bien si une précision exacte n'est pas nécessaire).

## 3) Other questions

- a)  Why is it important that a test set is independent to the training set in a machine learning system ?
    - Parce-qu'on veut mesurer la performance de notre modèle sur des données inconnues. On peut aussi avoir un problème d'overfitting, car les performances du test sera anormalement haute parce-qu'il a deja vu ces données. Le but d'un modèle est de performer sur des données qu'il n'a jamais vu et un training set indépendant simule le déployement du modèle dans le vrai monde => savoir si notre modèle va réalistiquement bien marcher.
- b) What's the meaning of the hat in the equation ^y = h(x) ?
    - Il indique qu'il s'agit de l'estimation, faite par notre modèle h(x), de la vraie valeur y pour une valeur inconnu x.
- c) What is the difference between machine learning and statistics ?
    - Statistiques : comprendre les relations entre les variables (ex : Est-ce que fumer cause le cancer du poumon ? Ces deux variables sont-elles liées ?)
    - Machine learning : But est de faire des prédictions sur des nouvelle données (ex : Avec les données de ce patient, va-t-il développer un cancer du poumon dans 5 ans ?)
- d) Is a detection system a regression or classification system ? Give an example.
    - Cela va dépendre de la sortie du système :
        - Si le modèle classe une entrée en catégories --> classification système (détecteur de spam email -> le système répond, spam ou not spam)
        - Si le modèle doit prédire une valeur continue --> regression problem (pour localiser un object dans une image, le système doit prédire les cordonnées exactes d'un box autour d'un objet)
